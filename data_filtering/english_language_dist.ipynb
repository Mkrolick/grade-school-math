```python
x = [x for x in range(0, len(data["count"]))]
y = data["count"].values
```

```python
# make a logathmic model fitting to x and y
# using ols
import statsmodels.api as sm
import numpy as np
import gmpy2
x = x
y = [float(gmpy2.log10(y)) for y in y]

#take log regression
x = sm.add_constant(x)
results = sm.OLS(y, x).fit()
results.summary()


# plot the data and the fitted curve
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(x, y, 'ko', label="Original Noised Data")
plt.plot(x, results.fittedvalues, 'r', label="OLS")
plt.title("Word Count")
plt.yscale("log")

```
